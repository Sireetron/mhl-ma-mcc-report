## Read Requests Per Second

```python
df = pd.read_excel('./Downloads/timeseriesd.xlsx',sheet_name='read')
```

```python
# fig, ax = plt.subplots(figsize=(9, 5))
sns.relplot(
    data=df, kind="line",
    x="timestamp", y="value",aspect=3
)
plt.xlabel("Day" ,fontsize=20)
plt.ylabel("Time (s)" ,fontsize=20)
plt.title("Read Requests Per Second" ,fontsize=20 ,weight='bold',loc='left')
plt.tick_params(labelsize=16)
```

## CUP

```python
df = pd.read_excel('./Downloads/timeseriesd.xlsx',sheet_name='CPU')
```

```python
fig, ax = plt.subplots(figsize=(13,5))
ax.plot(df['timestamp'],df['value']);
# make adjustments
ax.set_xlabel('Day',fontsize=20)
ax.set_ylabel('CPU Useage(%)',fontsize=20)

ax.text(x=0.1, y=1.2, s='Max CPU Useage(%)', fontsize=16, alpha=0.75, weight='bold', ha='right', va='bottom', transform=ax.transAxes)
ax.text(x=0.03, y=1.1, s='32 Core', weight='bold', fontsize=16, ha='right', va='bottom', transform=ax.transAxes)
plt.tick_params(labelsize=14)
h,l = ax.get_legend_handles_labels()
ax.legend(h[:4],l[:4], bbox_to_anchor=(1.05, 1), loc=2)

fig.tight_layout()
plt.show()
```
